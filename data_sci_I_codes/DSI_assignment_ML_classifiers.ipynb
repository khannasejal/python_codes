# SVM

```python
import pandas as pd
data_train = pd.read_excel("/content/Train.xlsx")
data_test = pd.read_excel("/content/Test.xlsx")

#splitting training and test data set into independent variables and dependent labels
X_train = data_train.iloc[:, 1:14]
X_test = data_test.iloc[:, 1:14]
y_train = data_train["PCE"] #true classification of train set
y_test = data_test["PCE"]  #true classification of test set

#Standardize the training and testing sets.
from sklearn import preprocessing

scalex = preprocessing.StandardScaler()
scalex.fit(X_train)
X_train = scalex.transform(X_train)
X_test = scalex.transform(X_test)
from sklearn import svm

#making the model by using algorithm
model = svm.SVC(kernel='rbf',C = 10.0,gamma=0.01)

#train the model
model.fit(X_train, y_train)

y_pred_test = model.predict(X_test) #testing the model on test set
y_pred_train = model.predict(X_train) #testing the model on train set
print("Accuracy for test set = ", accuracy_score(y_test, y_pred_test))  #method1, using accuracy_score()
print("Accuracy for train set = ", accuracy_score(y_train, y_pred_train))

score_train = model.score(X_train,y_train)  #method2, using model.score()
score_test = model.score(X_test,y_test)
print('Accuracy train = ', score_train)
print('Accuracy test = ', score_test)


```

# Random Forest Classifier

```python
import pandas as pd
data_train = pd.read_excel("/content/Train.xlsx")
data_test = pd.read_excel("/content/Test.xlsx")

#splitting training and test data set into independent variables and dependent labels
X_train = data_train.iloc[:, 1:14]
X_test = data_test.iloc[:, 1:14]
y_train = data_train["PCE"] #true classification of train set
y_test = data_test["PCE"]  #true classification of test set

from sklearn.ensemble import RandomForestClassifier
#build and optimize the model
model=RandomForestClassifier(n_estimators=12, max_depth=6, random_state=123)
model.fit(X_train, y_train)

score_train = model.score(X_train,y_train)
score_test = model.score(X_test,y_test)

print('Accuracy train = ', score_train)
print('Accuracy test = ', score_test)
```

```python
#feature importances
import pandas as pd
importances = model.feature_importances_
print(importances)
features = model.feature_names_in_
print(features)
```

```python
import pandas as pd

# Assuming you have trained a RandomForestClassifier named 'model'
importances = model.feature_importances_
print("All values of importances = ", importances)
feature_names = model.feature_names_in_  # Assuming you have column names
print("All feature names = ", feature_names)
# Create a DataFrame to store feature importances and names
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})

# Sort the DataFrame by importance in descending order
sorted_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Get the top 4 features and their importances
top_4_features = sorted_df.head(4)

# Print the top 4 features and importances
print("The top four features of the model are: ", top_4_features)
```

# Multiple Linear Regression


```python
train_data = pd.read_csv("/content/Train.csv")
test_data = pd.read_csv("/content/Test.csv")
```

```python
import numpy as np
X_train = train_data.iloc[:,2:15]
#axis=1: This parameter specifies that you are dropping a column (as opposed to a row).
#In pandas, axis=1 refers to columns, and axis=0 refers to rows.
y_train = train_data['PCE']
X_test = test_data.iloc[:,2:15]
#axis=1: This parameter specifies that you are dropping a column (as opposed to a row). In pandas, axis=1 refers to columns, and axis=0 refers to rows.
y_test = test_data['PCE']
#Standardize the training and testing sets.
from sklearn import preprocessing

scalex = preprocessing.StandardScaler()
scalex.fit(X_train)
X_train = scalex.transform(X_train)
X_test = scalex.transform(X_test)
#multiple linear regression for predicting the density.

from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)

from sklearn.metrics import r2_score
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)

print('R2 train = ', r2_score(y_train, y_pred_train))
print('R2 test = ', r2_score(y_test, y_pred_test))
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

print("RMSE for training data:", np.sqrt(mean_squared_error(y_train, y_pred_train)))
print("RMSE for test data:", np.sqrt(mean_squared_error(y_test, y_pred_test)))
```
